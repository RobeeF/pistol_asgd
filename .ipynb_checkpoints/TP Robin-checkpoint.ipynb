```python
def grad_quad_loss(theta,y,x):
    return np.mean(-x*(y-np.dot(x,theta.reshape(-1,1))),axis=0)

def grad_log_loss(theta,y,x):
    exp_term = np.exp(-y*np.dot(x,theta.reshape(-1,1)))
    return np.mean(- y*x*exp_term / (1. + exp_term),axis=0)
```

```python
def gd(theta_init, grad, n_iter=1000, step=0.99247, store_every=0.5, args=()):
    """Gradient descent algorithm."""
    theta = theta_init.copy()
    theta_list = []
    for i in range(n_iter):
        theta = theta - step * grad(theta, *args)
        if i % store_every == 0:
            theta_list.append(theta.copy())
    return theta, theta_list
```

### Non-averaged vs. averaged SGD

```python
def log_loss(theta,y,x,mu=0):
    ''' mu=0 to obtain the regular loss, mu>0 for penalized loss'''
    return np.mean(np.log(1+np.exp(-y*np.dot(x,theta.reshape(-1,1))))) + (mu/2)*np.linalg.norm(theta,ord=2)**2


def grad_log_loss_new(theta,y,x,mu=0):
    return np.mean(- y*x / (1. + np.exp(y*np.dot(x,theta.reshape(-1,1)))),axis=0) + mu*theta

loss = log_loss
grad = grad_log_loss_new
```

```python
def rate_1_sqrtn(mu,n):
    return (1/np.sqrt(n))

def rate_1_mun(mu,n):
    return 1/(mu*n)
```

```python
theta_init=np.zeros(100)
```

```python
def sgd(theta_init, y, x, grad, n_iter, step, mu=0, store_every=1, avg=False, cyclic=False):
    """Gradient descent algorithm."""
    theta = theta_init.copy()
    theta_avg = 0
    theta_list = []
    n = x.shape[0]

    for j in range(1,n_iter+1):
        chosen_idx = np.random.choice(n) if cyclic==False else (j-1)%n
        theta = theta - step(mu,j) * grad(theta, y[chosen_idx],x[chosen_idx,:], mu)
        if avg:
            theta_avg = (1-1/j)*theta_avg + (1/j)*theta
        if j % store_every == 0:
            if avg:
                theta_list.append(theta_avg.copy())
            else:
                theta_list.append(theta.copy())
    return (theta, theta_list) 
```

```python
n = X_train.shape[0]
n_iter = 20*n
step = rate_1_sqrtn
estimates = sgd(theta_init,Y_train,X_train,grad,n_iter=n_iter,step=step,store_every=100)[1]
estimates_avg = sgd(theta_init,Y_train,X_train,grad,n_iter=n_iter,step=step,store_every=100,avg=True)[1]
```

```python
test_error_sgd = []
test_error_sgd_avg = []
for coef in estimates:
    test_error_sgd.append(loss(coef, Y_test, X_test))
for coef in estimates_avg:
    test_error_sgd_avg.append(loss(coef, Y_test, X_test))
```

```python
plt.figure(figsize=(12,5))
plt.plot(test_error_sgd, lw=2)
plt.plot(test_error_sgd_avg, lw=2)
plt.title("Averaged vs. Non Averaged SGD")
plt.xlabel("Iterations (x100)")
plt.ylabel("Objective")
plt.legend(["Classic SGD", "Averaged SGD"])
plt.tight_layout()
plt.savefig('imgs2/cl_vs_avg_sgd.png', format='png', dpi=1000)
```

```python
def sag(theta_init, y, x, grad, epochs, step, epoch_len, store_every=1, mu=0.01):
    """Gradient descent algorithm."""
    theta = theta_init.copy()
    theta_list = []

    for i in range(epochs):
        all_grads = grad(theta, y,x, mu) # Compute all the gradients
        for new_i in range(epoch_len):
            chosen_idx = np.random.choice(n) 
            all_grads[chosen_idx] = grad(theta, y[chosen_idx],x[chosen_idx,:], mu)
            new_grad = np.mean(all_grads,axis=0)
            theta = theta - (1/3000)*new_grad
            
            if new_i % store_every == 0: # Attention quand d√©sindente
                theta_list.append(theta.copy())
    return theta, theta_list
```

```python
def smooth_hinge(y,y_pred):
    loss = 0 if abs(y-y_pred)>=1
```

```python
def grad_quad_loss(y,y_pred):
    return np.mean(y-y_pred,axis=0) 
```

```python
def quad_loss(y,y_pred):
    return np.mean((y-y_pred)**2,axis=0) 
```

```python
def gaussian_kernel(x, y, gamma=0.04):
    return np.exp(-gamma*np.linalg.norm(x-y)**2)
```

```python
def compute_kernel(X):
    ''' From mblondel percepetron Github'''
    n_samples = X.shape[0]
    K = np.zeros((n_samples, n_samples))
    for i in range(n_samples):
        for j in range(n_samples):
            K[i,j] = gaussian_kernel(X[i], X[j])
    return K 
```

```python
nb_obs = 2000
```

```python
kernel = compute_kernel(X_train[:nb_obs,:])
```

```python
def pistol(y, x, grad, n_iter, L, mu=0, store_every=1, b=1):
    """Gradient descent algorithm."""
    n, d = x.shape
    a = 0.25
    beta = np.sqrt(2*a*L*n) # Make only one pass over the data
    alpha = a*L
    
    f = np.zeros((n_iter,n))
    g = np.zeros((n_iter, n))
    
    for t in range(1, n_iter):

        f[t] = g[t-1]*(b/alpha)*np.exp(gaussian_kernel(g[t-1], np.zeros(n))**2/(2*alpha)) # Change to Norm K
        y_pred = f[t,t%n] # f[t,t%n] might me better
                
        s_t =  grad(y_pred, y[t])
        # Updates
        g[t] = g[t-1] - s_t*kernel[t]
        #print("Kernel norm: ", gaussian_kernel(kernel[t], np.zeros(n)))
        alpha = alpha + a*np.abs(s_t)*gaussian_kernel(kernel[t], np.zeros(n)) # Check norm
    return f#np.mean(f,axis=0)
```

```python
n_iter = nb_obs
f_history = pistol(Y_train[:nb_obs], X_train[:nb_obs], grad=grad_quad_loss, n_iter=n_iter, L=40)
```
